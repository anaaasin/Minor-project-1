```python
import networkx as nx
import matplotlib.pyplot as plt


graph = nx.Graph()   # this creates an undirected graph

graph.add_node(1)                     # add single node
graph.add_nodes_from([2,3,4])         # add multiple nodes

# adding a node with attribute
graph.add_node(5, role="human")

print("Nodes in graph:")
print(graph.nodes(data=True))

graph.add_edge(1,2)

graph.add_edges_from([
    (1,3),
    (2,4),
    (3,4),
    (2,5)
])

# edge with weight
graph.add_edge(1,4, weight=5)

print("\nEdges in graph:")
print(graph.edges(data=True))

print("\nTotal number of nodes:", graph.number_of_nodes())
print("Total number of edges:", graph.number_of_edges())

print("\nNeighbors of node 1:")
print(list(graph.neighbors(1)))

print("\nDegree of node 1:", graph.degree(1))

print("\nDegree of all nodes:")
for node, degree in graph.degree():
    print("Node", node, "has degree", degree)

print("\nAdjacency List:")
for line in nx.generate_adjlist(graph):
    print(line)

print("\nAdjacency Matrix:")
matrix = nx.adjacency_matrix(graph).todense()
print(matrix)

path = nx.shortest_path(graph, source=1, target=4)
print("\nShortest path from 1 to 4:", path)

length = nx.shortest_path_length(graph, source=1, target=4)
print("Shortest path length:", length)

print("\nBFS traversal starting from node 1:")
bfs_nodes = list(nx.bfs_tree(graph, source=1))
print(bfs_nodes)


print("\nDFS traversal starting from node 1:")
dfs_nodes = list(nx.dfs_tree(graph, source=1))
print(dfs_nodes)

print("\nIs graph connected?")
print(nx.is_connected(graph))


print("\nConnected components:")
components = list(nx.connected_components(graph))
print(components)


graph.remove_edge(1,2)
print("\nEdges after removing (1,2):")
print(list(graph.edges()))

graph.remove_node(5)
print("\nNodes after removing node 5:")
print(list(graph.nodes()))


nx.draw(graph, with_labels=True, node_size=2000, node_color="lightblue")

plt.title("Simple NetworkX Graph")
plt.show()
```
